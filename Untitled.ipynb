```python
#Aula 04
#LISTA indicada por dois [] 
carros = ['bmw','toyota','fiat','vw','gm','audi','ferrari']
print(carros)
print(f'Os carros da loja são: {carros}') #Lista original
print(carros[0]) #posição zero / sempre começa em 0
print(carros[4].title()) #primeira letra maiuscula
print(carros[3].upper()) #tudo maiusculo
print(carros[-1]) #último elemento na lista
carros[2] = 'chery'#Substituir
print(carros)
carros.append('honda') #Adicionar um item no final da lista
print(carros)
carros.insert(2,'Jack') #Insere em uma posição específica 
print(carros)
del carros[3] #exclui o valor que ta na posição três
print(carros)
carros.pop()
print(carros) #Remove o último item da lista
carros.pop(1)
print(carros)
carros.remove('audi') #Remove o item procurado
print(carros)  
carros.sort() #Organizar a lista
print(carros)
carros.sort(reverse=True)
print(carros)
print(sorted(carros)) #organiza de forma temporária
print(len(carros)) #Tamanho da lista
```
